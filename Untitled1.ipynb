```python
def create_dataset(dataset, look_back=3):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i:(i + look_back)]
        dataX.append(a)
        dataY.append(dataset[i + look_back])
    return np.array(dataX), np.array(dataY)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Input,LSTM,Dropout,Dense



# create and fit the LSTM network

model = keras.Sequential()
model.add(LSTM(260,input_shape=(5,3),return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(260,input_shape=(5,3),return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(260,input_shape=(5,3),return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(260,input_shape=(5,3),return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(5,activation='linear'))
model.compile(optimizer='rmsprop',loss='mean_squared_error',metrics=['accuracy'])

history = model.fit(train_X, train_Y, epochs=5, batch_size=100, verbose=2)
```

```python
def show_curve_plot(x, length, first_5_eigenworms, start=0, title="Plot", vmin=0, vmax=1, bar_pos=10):
    '''
    Function to plot curvature/time of an input array x (shape - (N,5)).
    '''
    START=start
    NUM_FRAMES=length
    x= np.array(x)
    q = np.sum(x[0].reshape(5,1)*first_5_eigenworms,axis=0)
    for j in range(NUM_FRAMES - 1):
        new_row = np.sum(x[START+j+1].reshape(5,1)*first_5_eigenworms,axis=0)
        q = np.vstack((q,new_row))
    plt.rcParams['figure.figsize'] = 20, 20
    plt.title(title)
    plt.yticks(np.arange(0, 100, 20.0))
    plt.xticks(np.arange(0, length, 100.0))
    plt.ylabel('Body Position (u)')
    plt.xlabel('Frame Number (16FPS)')
    print("QMin: " + str(q.min()) + " QMax: " + str(q.max()) + " QMean: " + str(q.mean()))
    plt.imshow(q.transpose(),cmap='viridis',vmin=vmin, vmax=vmax)
#     plt.axvline(x=bar_pos, ymin=0, ymax=100, linewidth=1, color='k')
    plt.gca().invert_yaxis()
    ax = plt.gca()
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)
    plt.colorbar(cax=cax)
    plt.show()
```

```python
def plot_coefficient_reconstruction(model, X, Y):
    plt.clf()
    fig, axarr = plt.subplots(2, 2)
    for coeff_idx in range(5):
        if(coeff_idx == 0):
            i, j = 0,0
        elif(coeff_idx == 1):
            i, j = 0,1
        elif(coeff_idx == 2):
            i, j = 1,0
        elif(coeff_idx == 3):
            i, j = 1,1
        else:
            break
        s=1.
        lw = .8
        MAX = 100000
        #print(X.shape)
        prediction_raw_sample = model.predict(X)[0:MAX][:,coeff_idx]
        Y_raw_sample = Y[0:MAX][:,coeff_idx]
        predictions = prediction_raw_sample

        #print(predictions.shape)
        #plt.scatter(np.arange(0,Y_raw_sample.shape[0],1), (predictions), s=s, label="Prediction")
        #plt.scatter(np.arange(0,Y_raw_sample.shape[0],1), Y_raw_sample[0:], s=s, label="Real")
        axarr[i,j].plot((predictions), label="Predicted", c="red", linewidth=lw)
        axarr[i,j].plot(Y_raw_sample[:], label="Data", linewidth=lw)
        axarr[i,j].title.set_text("Mode " + str(coeff_idx+1))
        axarr[i,j].set_ylim([0,1])
        axarr[i,j].set_xlim([0,length])
        axarr[i,j].set_xlabel("Frame Number (16FPS)")
        axarr[i,j].set_ylabel("Coefficient")
        axarr[i,j].axvline(x=start_of_test, ymin=0, ymax=100, linewidth=1, color='k')


#         axarr[i,j].legend(loc="upper right")
        axarr[i,j].legend(fontsize=10,bbox_to_anchor=(1, 1),loc='center')
#     plt.figlegend( )
    plt.rcParams['figure.figsize'] = 5,10
    plt.tight_layout(pad=3.0, w_pad=5.5, h_pad=1.0)
    plt.suptitle("Eigemode Reconstruction")
    plt.show()
plot_coefficient_reconstruction(model, test_X[start:start+length], test_Y[start:start+length])
```
